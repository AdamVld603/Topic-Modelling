```python
# Imports

from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer


import numpy as np
import os
import pandas as pd
import re
import string
import regex

import spacy
from spacy import displacy

import gensim
from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel
from gensim.models.wrappers import LdaMallet
from gensim.corpora import Dictionary
import pyLDAvis.gensim

import time
```

```python
# --- Read Cleaned and Preprcessed data:

# Provide Location of cleaned/preprcessed csv's:
data_location_processed = '.\\data_cleaned_processed\\' #Change path to where original .csv's are
my_directory_processed = os.listdir(data_location_processed) 
print(my_directory_processed)
```

```python
# === Create Topic Models for each year:

start_time_total = time.time()

for y in my_directory_processed:  
    start_time_year = time.time()
    news_ = pd.read_csv(data_location_processed +y, encoding='utf-8', dtype=str) #header = None)
    #print(news_.head())

    # Removesw null values:
    #news_ = news_[news_['article'].notnull()] 
    news_ = news_.dropna(how='any',axis=0) 
    
    # ---------- Combine "article" and "title" columns: [OPTIONAL]
    #news_["article"] = news_["title"].map(str) + " " + news_["article"]#erdan
    #print(news_.head())   
    
    # Tokenize each document
    news_ = list(news_["article"])
    news_tokens_ = [ x.split() for x in news_]
    #print(news_tokens_[1][0:5])
    
    # ------- bigram: [OPTIONAL]
    print("Doing Bigrams ... ")
    bigram_ = gensim.models.Phrases(news_tokens_)
    #bigram = gensim.models.phrases.Phraser(news_tokens_) # For a faster implementation
    texts_ = [bigram_[line] for line in news_tokens_]
    #print(texts_[0])

    # ----- Convert our processed/cleaned news data into a "bag of words" model
    print("Doing bag_of_words ... ")
    dictionary_ = Dictionary(texts_)#(news_tokens)
    corpus_bow_ = [dictionary_.doc2bow(text) for text in texts_]# news_tokens]
    #print(corpus_bow_[1][0:5])    
    #dictionary.get(21)
    #len(corpus_bow)
    
    title_ = y
    #title_ = title_.replace('_cleaned_', '')
    title_ = title_.replace('.csv', '')
    path_cleaned =  '.\\data_models\\'
    print("*"*100)
    
    # ======= Topic models
    #---------
    # -------------- LSI (Latent semantic analysis)
    print("="*100)
    print("Processing LSI: ", title_, " ...")
    lsimodel = LsiModel(corpus=corpus_bow_, num_topics=20, id2word=dictionary_)
    #print(lsimodel.show_topics(num_topics=3, num_words = 30))  # Showing only the top 3 topics
    lsimodel.save(path_cleaned + 'lsi\\'+'lsi'+title_) #save model
    #---------
    # -------------- LDA (Latent Dirichlet allocation)
    print("="*100)
    print("Processing LDA: ", title_ , " ...")
    ldamodel = LdaModel(corpus=corpus_bow_, num_topics=20, id2word=dictionary_)
    #print(ldamodel.show_topics(num_topics=3, num_words = 30))  # Showing only the top 3 topics
    ldamodel.save(path_cleaned + 'lda\\'+'lda'+title_) #save model
    #---------
    # -------------- HDP (Hierarchical Dirichlet Process)
    print("="*100)
    print("Processing HDP: ", title_, " ...")
    hdmodel = HdpModel(corpus=corpus_bow_, id2word=dictionary_)
    #hdmodel.show_topics()  # Showing topics 
    hdmodel.save(path_cleaned + 'hdp\\'+'hdp'+title_) #save model
    end_time_year = time.time()
    print("-"*100)
    print("Total time for --> ", title_ , " : ", end_time_year-start_time_year)
    
end_time_total = time.time()
print("*"*100)
print("Total time: ", end_time_total-start_time_total)
```
